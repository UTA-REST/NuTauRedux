```python
from icecube import dataio
from icecube import dataclasses

taufile = dataio.I3File('/data/sim/IceCube/2011/filtered/level2/neutrino-generator/10099/00000-00999/Level2_nugen_nutau_IC86.2011.010099.000002.i3.bz2')
geofile = dataio.I3File('/data/sim/IceCube/2011/filtered/level2/neutrino-generator/10099/00000-00999/GeoCalibDetectorStatus_IC86.55697_corrected_V2.i3.gz')
g_frame = geofile.pop_frame()
geo=g_frame['I3Geometry']
c_frame = geofile.pop_frame()
calibration = c_frame["I3Calibration"]
d_frame = geofile.pop_frame()
detector = d_frame["I3DetectorStatus"]

#for i,j in frame['InIceRawData']:
#   print(dataclasses.atwd_sampling_rate(0,detector.dom_status[i],calibration.dom_cal[i]))
#dom_pos= [j.position for i,j in geo.omgeo]
```

```python
InIceRawData=frame['InIceRawData']
OfflinePulses=frame['OfflinePulses']
MCTree=frame['I3MCTree']
```

```python
TauPos=dataclasses.I3Position(MCTree[1].pos)
print(TauPos)
```

```python
FoundTau=0
for i in MCTree:
   # print(i.type)
    if(i.type_string=="TauMinus" or i.type_string=="TauPlus"):
        FoundTau=1
        TauPos=i.pos
        TauLength=i.length
        TauTime=i.time
        print("TauPos", TauPos)
        print(i)
    if((i.type_string=="NuTau" or i.type_string=="NuTauBar") and FoundTau==1):
        NuTauPos=i.pos
        print("NuTauPos",NuTauPos)
```

```python
OfflinePulses=frame['OfflinePulses']
COG=dataclasses.I3Position(0,0,0)
TotalCharge=0
for pulse in OfflinePulses:
    omkey=pulse[0]
    position=geo.omgeo[omkey].position
    ChargeSum=sum(j.charge for j in pulse[1])
    COG+=position*ChargeSum
    TotalCharge+=ChargeSum
COG=(1./TotalCharge)*COG
print(COG)
```

```python
sampletime=1./0.3
c=0.3
xdiffs=[]
ydiffs=[]
zdiffs=[]
corrs=[]
nanosecdict=[{},{},{},{}]
TimeThresh=100
for x in InIceRawData:
    count=0
    omkey=x[0]
    position=geo.omgeo[omkey].position
    diff=(position-COG)
    absdiff=pow(diff[0]*diff[0]+diff[1]*diff[1]+diff[2]*diff[2],0.5)
    TimeCorr=absdiff/(sampletime*c)
    corrs.append(TimeCorr)
  #  print(absdiff,TimeCorr)
    if(TimeCorr<TimeThresh):
        for j in x[1]:
            TimeStartsamp=int(j.time/sampletime-TimeCorr)

            for atwd in range(0,len(j.raw_atwd)):
                for sampleno in range(0,len(j.raw_atwd[atwd])):
                    sample=j.raw_atwd[atwd][sampleno]
                    index=int(TimeStartsamp+sampleno)
                    if not nanosecdict[atwd].has_key(index):
                        nanosecdict[atwd][index]=0
                    nanosecdict[atwd][index]=nanosecdict[atwd][index]+j.raw_atwd[atwd][sampleno]
 
```

```python
for i in nanosecdict:
    times=i.keys()
    amps=i.values()
    pylab.plot(times,amps,'.')
    if(len(amps)>0):
        pylab.plot([(TauTime)/sampletime,(TauTime)/sampletime],[0,max(amps)],color='red',linestyle='--')
        pylab.plot([(TauTime+TauLength/c)/sampletime,(TauTime+TauLength/c)/sampletime],[0,max(amps)],color='red',linestyle='--')
        pylab.show()


pylab.plot(corrs,'.')
pylab.ylabel("Size of time correction")
pylab.ylabel("Pulse ID")
pylab.show()
```
