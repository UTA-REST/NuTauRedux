```python
from __future__ import division, print_function
import numpy, pylab, math
import icecube
from icecube import dataio
from icecube import dataclasses
from icecube import CascadeVariables
import os
%matplotlib inline
```

```python
InFile= dataio.I3File("./output_sig.i3")
```

```python
todo=20
count=0
diffset=20
InFile.rewind()

while InFile.more():
    pylab.figure(figsize=(5,5))
    count+=1
    fr=InFile.pop_physics()
    SumWF=fr["SumWF"]
    DecWF=fr["DecWF"]
    CumWF=numpy.cumsum(DecWF*(numpy.array(DecWF)>0))
    nums=numpy.arange(0,len(SumWF))
    pylab.fill_between(nums,0,numpy.array(SumWF)/max(SumWF),linewidth=2,color='blue',alpha=0.2)
    pylab.plot(nums+diffset,numpy.array(DecWF)/max(DecWF),linewidth=2,color='red')
#    pylab.plot(nums+diffset,numpy.array(CumWF)/max(CumWF),linewidth=2,color='black')


    pylab.xlim(0,400)
  #  pylab.ylim(0,1)
    pylab.show()
    if(count==todo):
        break
        
```
